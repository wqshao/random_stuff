```python
import string
import itertools
import numpy as np

from copy import copy

inputs = []
with open("day19.txt", "r") as f:
    inputs = list(map(int, f.readline().split(',')))

p1 = 0
p2 = 0

POS = 0
IMM = 1
REL = 2
mode = POS
```

```python
def run(reg, value):
    rel_base = 0
    i = 0
    #output = []
    while True:
        ins = reg[i]
        mp1 = int(ins / 100) % 10
        mp2 = int(ins / 1000) % 10
        mp3 = int(ins / 10000) % 10
        ins = ins % 100
        
        #print(ins)
        if ins == 99:
            break
        if mp1 == IMM: r1 = i+1
        elif mp1 == POS: r1 = reg[i+1]
        elif mp1 == REL: r1 = reg[i+1]+rel_base
        if mp2 == IMM: r2 = i+2
        elif mp2 == POS: r2 = reg[i+2]
        elif mp2 == REL: r2 = reg[i+2]+rel_base   
        try:
            if mp3 == IMM: r3 = i+3
            elif mp3 == POS: r3 = reg[i+3]
            elif mp3 == REL: r3 = reg[i+3]+rel_base  
        except IndexError:
            pass # if this happens, you don't need it
        #print(r1, r2, r3)
        
        if ins == 1:
            reg[r3] = reg[r1] + reg[r2]
            i += 4
        elif ins == 2:
            reg[r3] = reg[r1] * reg[r2]
            i += 4
        elif ins == 3:
            reg[r1] = yield "input value" 
            i += 2
        elif ins == 4:
            value = reg[r1]
            i += 2
            yield value 
            #output.append(value)
            #print(value)

            #if reg[i] % 100 == 99:
            #    print(value)
            #else:
            #    print(value)
        elif ins == 5:
            if reg[r1]:
                i = reg[r2]
            else:
                i += 3
        elif ins == 6:
            if reg[r1] == 0:
                i = reg[r2]
            else:
                i += 3
        elif ins == 7:
            reg[r3] = 1 if reg[r1] < reg[r2] else 0
            i += 4
        elif ins == 8:
            reg[r3] = 1 if reg[r1] == reg[r2] else 0
            i += 4
        elif ins == 9:  
            #print(rel_base)
            rel_base += reg[r1]
            i += 2
        else:
            print('failed')
            break
    
    return 
```

```python
def check(i, j):
    reg = copy(inputs)
    orig_length = len(reg)
    reg[orig_length:1000000] = [int(0)] * (1000000 - orig_length)

    intcode = run(reg, None)
        
    next(intcode)
    intcode.send(i)
    reply = intcode.send(j)
    return reply
```

```python
from __future__ import print_function 

def print_grid(grid, x=0, y=0):
    keys = grid.keys()
    minx = min(x[0] for x in keys)
    miny = min(x[1] for x in keys)
    maxx = max(x[0] for x in keys)
    maxy = max(x[1] for x in keys)

    for y in range(miny, maxy + 1):
        for x in range(minx, maxx + 1):
            print(grid.get((x, y), " "), end="")
        print()

    return "Printed"
```
